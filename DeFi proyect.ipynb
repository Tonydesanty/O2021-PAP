```python
import numpy as np
from pandas_datareader import data as wb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance
import warnings
import scipy.stats
import pylab
warnings.filterwarnings("ignore")
sns.set()
```

```python
def get_tickets_close(tickets):
    info = yfinance.download(tickers = tickets, interval ='1d',start='2016-01-01', end ='2021-09-09', groub_by = 'ticker',auto_adjust = True, treads = True)['Close']
    return info
```

```python
def getExpectedReturn(df, price_col_name, annualised=True, annualise_method='sophisticated'):
    """
    Returns the expected return of a stock given price data.
    """

    # Calculate returns of prices
    returns = df[price_col_name].pct_change(1)
    
    # Calculate the expected return using the mean method
    expected_return_daily = returns.mean()
    
    if annualised:
        if annualise_method == 'sophisticated':
            expected_return_annual = ((1 + expected_return_daily) ** 250) - 1
        elif annualise_method == 'crude':
            # Crude method
            expected_return_annual = expected_return_daily * 250

        return expected_return_annual
    
    else:
        return expected_return_daily
```

```python
tickets = ["BTC-USD","LTC-USD","ETH-USD", "XRP-USD", "XMR-USD", "DASH-USD", "XLM-USD"]
crypto = get_tickets_close(tickets)
crypto.tail()
```

```python
print("Bitcoin's returns since 2016 is:",str(getExpectedReturn(crypto,'BTC-USD')))
print("Dash's returns since 2016 is:",str(getExpectedReturn(crypto,'DASH-USD')))
print("Ether's returns since 2016 is:",str(getExpectedReturn(crypto,'ETH-USD')))
print("Litecoins returns since 2016 is:",str(getExpectedReturn(crypto,'LTC-USD')))
print("Stellar's returns since 2016 is:",str(getExpectedReturn(crypto,'BTC-USD')))
print("Monero's returns since 2016 is:",str(getExpectedReturn(crypto,'XMR-USD')))
print("XRP's returns since 2016 is:",str(getExpectedReturn(crypto,'BTC-USD')))
```

DeFi solved these problems with cryptocurrencies named **"Stablecoins".**

The purpose of stablecoins is to eliminate cryptocurrencies volatility, this with the goal of approaching its value to traditional currencies.

Stablecoins have the same characteristics of cryptocurrencies, they can be sent over the world using internet, without an intermediary. Another important characteristic of the coins is that they are secured in cryptography so no one can duplicate your money or eliminate your money.

The most famous stablecoin is **DAI**, the reason is that the value of these stablecoin is very similir to the dollar, and it´s accepted across **DAPS**(decentralized applications).


There are different kinds of stablecoins with different purposes, this with the purpose of stable the volatility of different kind of assets or the type of usage that the user is willing to make.

**FIAT BACKED**: This type of stablecoins aims to protect the volatility of the cryptocurrencies, you buy a stablecoin that is related with a currency like dollar, and like this you completely eliminate crypto volatility. A issue that have fiat backed stablecoins is that is centralizaed, someone must deliver the tokens.

**Cryptobacked stablecoins:"** This type of stablecoins are used to protect the volatility of certain assets. The value of this kind of stablecoins derivates from the value of the crypto (collateral). Tthis stablecoin is fully dezantralized but less stable than *fiat backed* stablecoin.

**Precious metals stablecoins:"** Are very similiar to fiat backed stablecoins but it´s value depends of a precious metal like gold.
